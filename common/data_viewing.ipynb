# Data Viewing
This file is used to visualize the covid data over time.

### Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import pickle
```

### Import the list of regions

```python
with open('../data/region_list.pkl', 'rb') as file:
    region_names = pickle.load(file)
```

### Import the data as DataFrames

```python
# Lists of all regional features
daily_counts = []
daily_rates = []
daily_rates_3dra = []
daily_rates_7dra = []

# Import counts of covid per day for each region
for file in Path('../data/dataframes/daily_counts').glob('*'):
    if file.name != 'Total.pkl': daily_counts += [pd.read_pickle(file)]
    else: daily_counts_total = pd.read_pickle(file)

# Import rates of covid per day for each region
for file in Path('../data/dataframes/daily_rates').glob('*'):
    if file.name != 'Total.pkl': daily_rates += [pd.read_pickle(file)]
    else: daily_rates_total = pd.read_pickle(file)

# Import a 3 day rolling average of covid rates per day for each region
for file in Path('../data/dataframes/3_day_rolling_average_rates').glob('*'):
    if file.name != 'Total.pkl': daily_rates_3dra += [pd.read_pickle(file)]
    else: daily_rates_3dra_total = pd.read_pickle(file)

# Import a 7 day rolling average of covid rates per day for each region
for file in Path('../data/dataframes/7_day_rolling_average_rates').glob('*'):
    if file.name != 'Total.pkl': daily_rates_7dra += [pd.read_pickle(file)]
    else: daily_rates_7dra_total = pd.read_pickle(file)
```

### Plot each region's case count over time

```python
# Plot each region's covid cases over time
for region in daily_counts: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot the total case counts over time

```python
# Plot each region's covid cases over time
sns.lineplot(x='index', y='confirmed', data=daily_counts_total)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot each region's cases per day over time

```python
# Plot each region's covid cases over time
for region in daily_rates: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot each region's 3 day rolling average of cases per day over time

```python
# Plot each region's covid cases over time
for region in daily_rates_3dra: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot each region's 7 day rolling average of cases per day over time

```python
# Plot each region's covid cases over time
for region in daily_rates_7dra: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot the total rates over time

```python
# Plot each region's covid cases over time
sns.lineplot(x='index', y='confirmed', data=daily_rates_total)
sns.lineplot(x='index', y='confirmed', data=daily_rates_3dra_total)
sns.lineplot(x='index', y='confirmed', data=daily_rates_7dra_total)

# Set the plot legend
plt.legend(labels=['Daily Rates', 'Daily Rates 3 Day Rolling Average',
                   'Daily Rates 7 Day Rolling Average'])

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```
