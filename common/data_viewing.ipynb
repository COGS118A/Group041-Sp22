# Data Viewing
This file is used to visualize the covid data over time.

### Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import pickle
```

### Import the list of regions

```python
with open('../data/region_list.pkl', 'rb') as file:
    region_names = pickle.load(file)
```

### Import the data as DataFrames

```python
# A list of all regional dataframes
regions_absolute = []
regions_relative = []

# Import all regional data about total infection counts
for file in Path('../data/dataframes/absolute').glob('*'):
    regions_absolute += [pd.read_pickle(file)]

# Import all data about infection rates
for file in Path('../data/dataframes/relative').glob('*'):
    regions_relative += [pd.read_pickle(file)]
```

### Plot each region's case count over time

```python
# Plot each region's covid cases over time
for region in regions_absolute: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```

### Plot each region's cases per day over time

```python
# Plot each region's covid cases over time
for region in regions_relative: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()

```
