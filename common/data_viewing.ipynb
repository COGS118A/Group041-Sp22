# Data Viewing
This file is used to visualize the covid data over time.

### Imports

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import pickle
```

### Import the list of regions

```python
with open('../data/region_list.pkl', 'rb') as file:
    region_names = pickle.load(file)
```

### Import the data as DataFrames

```python
# A list of all regional dataframes
regions = []

# Make dataframes from each region's pickle file
for file in Path('../data/dataframes').glob('*'):
    regions += [pd.read_pickle(file)]
```

### Plot each region's case count over time

```python
# Plot each region's covid cases over time
for region in regions: sns.lineplot(x='index', y='confirmed', data=region)

# Set the plot legend
plt.legend(labels=region_names)

# Change the plot size
fig = plt.gcf()
fig.set_size_inches(16, 10.5)

# Display the plot
plt.show()
```
